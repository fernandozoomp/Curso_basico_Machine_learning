```python
from os import listdir
from os.path import isfile, join

def get_existent_text_files(folder, name_only=False):

    files_only = [f for f in listdir(folder) if isfile(join(folder, f))]
    txts_only = [f for f in files_only if '.txt' in f]
    if name_only:
        names_only = [f.replace('.txt', '') for f in txts_only]
        return names_only
    else:
        return [join(folder, c) for c in txts_only]
```

```python
def remove_tag(text):
    first_col = text.index(']')
    return text[first_col + 1:]
```

```python
folders = ['concessÆo_da_ordem', 'concessÆo_da_ordem_de_of¡cio',
    'concessÆo_parcial_da_ordem', 'denegaÆo_da_ordem']
```

```python
from tqdm import tqdm

texts = []
tags = []

ROOT = join('datasets', 'entrada')

for folder in tqdm(folders):
    folder_to_read = join(ROOT, folder)
    current_files = get_existent_text_files(folder_to_read)

    for cfile_path in current_files:
        ctext = open(cfile_path, 'r', encoding='utf-8').read()
        texts.append(remove_tag(ctext))
        tags.append(folder)
```

```python
len(texts)
```

```python
all_tags = list(set(tags))
```

```python
tags_n = [all_tags.index(t) for t in tags]
```

```python
from sklearn.model_selection import train_test_split

texts_train, texts_test, tags_train, tags_test = train_test_split(
texts, tags_n, test_size=0.2, stratify=tags_n, shuffle=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(texts_train)
X_test = vectorizer.transform(texts_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, tags_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

prediction = model.predict(X_test)

cm = confusion_matrix(tags_test, prediction)
p_score = precision_score(tags_test, prediction, average=None)
r_score = recall_score(tags_test, prediction, average=None)

print('Confusion matrix')
print(cm)
print('Precision score', p_score)
print('Recall score', r_score)
print('All classes', all_tags)
```
