<h3>List comprehension</h3>
<br>
Listas em python podem ser transformadas em outras listas de forma bastante simples

```python
a = [1,2,3,4]
b = [2*x for x in a] #Crie uma lista onde os elementos são o dobro dos correspondentes em a
print(b)
```

List comprehensions podem filtrar dados tambem

```python
#Crie uma lista onde os elementos são o dobro dos correspondentes em a, mas apenas com os pares
c = [2*x for x in a if x % 2 == 0]
print(c)
```

List comprehensions podem ser aplicados de forma aninhada

```python
list1 = [3, 7, 60, 20, 1]
thresholds = [5, 30]
filtered_lists = [[y for y in list1 if y < t] for t in thresholds]
print(filtered_lists)
```

Listas podem ser ordenadas a partir do comando sorted

```python
#Funcoes podem retornar mais de um valor e ter parametros nomeados com valores default
def incr(x, also_decrement=False):
    if not also_decrement:
        return x + 1
    else:
        return x + 1, x - 1

print(incr(1))
print(incr(1, also_decrement=True))
```

<h1>Atividade 1: Faça uma função que retorne uma lista dos n números da sequência de Fibonacci</h1>
    
1, 1, 3, 4, 7, 11, .....

+ fib(x) = fib(x - 1) + fib(x - 2)
+ fib(0) = 1
+ fib(1) = 1

```python
def fib(x):
    if x == 0 or x == 1:
        return 1
    else:
        return fib(x - 1) + fib(x - 2)

def fib_list(n):
    return [fib(x) for x in range(n)]

print(fib_list(10))
```

<h1>Atividade 2: Para o texto example_text.txt, crie um dicionário que relacione cada palavra à sua frequência no texto</h1>

```python
example_text = open('datasets/example_text.txt', 'r').read()
print(example_text[:100])
```

```python
freq_dist = {}
splitted_text = example_text.split(' ')

#print(splitted_text[:10])

for word in splitted_text:
    if word in freq_dist:
        freq_dist[word] += 1
    else:
        freq_dist[word] = 1

#print(freq_dist)
freq_dist_list = list(freq_dist.items())
sorted_freq_dist = sorted(freq_dist_list, key=lambda x: x[1], reverse=True)
print(sorted_freq_dist)
```

```python
#Alguns outros testes

from tqdm import tqdm

letters = ['a', 'b', 'c', 'd', 'e']

all_names = []

for letter in tqdm(letters):
    url = 'https://www.dicionariodenomesproprios.com.br/bebes/%s/' % (letter)
    raw_html = urllib.request.urlopen(url).read()

    soup = Soup(raw_html)
    all_links = soup.findAll('a')
    links_with_class = [L for L in all_links if L.has_attr('class')]
    class_name = [L for L in links_with_class if L['class'][0] == 'name-link']
    names = [L.text for L in class_name]
    all_names += names

print(all_names)
```

```python
#Outro teste

url = 'http://www.sport-histoire.fr/pt/Geografia/Paises_por_ordem_alfabetica.php'
custom_headers = { 'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)' }

req_obj = urllib.request.Request(url, headers=custom_headers)
response = urllib.request.urlopen(req_obj)

raw_html = response.read()

soup = Soup(raw_html)
all_tds = soup.findAll('td')
all_countries = []

for td in tqdm(all_tds):
    all_a = td.findAll('a')
    if len(all_a) > 0:
        all_countries.append(all_a[0].text)
all_countries
```
